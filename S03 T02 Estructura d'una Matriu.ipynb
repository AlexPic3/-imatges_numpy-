### Exercici 4
* Mostreu-me amb exemples de diferents matrius, la regla fonamental de Broadcasting que diu : "les matrius es poden transmetre / broadcast si les seves dimensions coincideixen o si una de les matrius té una mida d'1".

Es posible realizar operaciones en ndarrays de diferentes tamaños. 
En algunos casos numpy puede transformar estos ndarrays automáticamente
de forma que todos tienen la misma forma. Esta conversión automática se
llama broadcasting.

***The Broadcasting Rule\
In order to broadcast, the size of the trailing axes for both arrays in an operation must either be the same size or one of them must be one.***

```python
# Este codigo dara el error:
# ValueError: operands could not be broadcast together with shapes (4,3) (2) 
from numpy import array
a = array([[ 0, 0, 0],
            [10,10,10],
            [20,20,20],
            [30,30,30]])
b = array([1,2])
a + b
```

```python
from numpy import array
a = array([[ 0, 0, 0],
            [10,10,10],
            [20,20,20],
            [30,30,30]])
b = array([1,2,3])
a + b
```

```python
from numpy import array
a = array([[ 0, 0, 0],
            [10,10,10],
            [20,20,20],
            [30,30,30]])
b = array([3])
a + b
```

```python
# Matrix multiplication not work with a*b 
# ValueError: operands could not be broadcast together
from numpy import array
a = array([[ -1, 1, 3],
            [2,5,-7]])
b = array([[2,-6],
            [1,0],
            [-2,8]])
c = np.matmul(a,b)
print ("AB =", c)
d = np.matmul(b,a)
print ("BA =", d)
```
